```python
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision import datasets
from PIL import Image
import os
```

```python
# 创建一个 12x12 大小的画布，用于绘制图像
fig, ax = plt.subplots(figsize=(12, 12))
```

```python
# 去掉了图像显示中的坐标轴的刻度，使得显示的图片更加清晰，专注于图片本身
ax.set_xticks([]); ax.set_yticks([])
```

```python
image_tensors = []

trans = transforms.Compose([
    transforms.ToTensor()
])

image_folder = '../../dataset/cvimages'
# 列出文件夹中的图片文件，并只选择前 60 张
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', 'tif'))]
image_files = sorted(image_files)[:60]  # 按字母顺序排序，并选前60张

# 读取并展示前60张图片
for image_name in image_files:
    image_path = os.path.join(image_folder, image_name)
    
    # 使用PIL打开图片
    img = Image.open(image_path)
    image_tensors.append(trans(img))
    
    # # 展示图片
    # img.show()

ax.imshow(make_grid(image_tensors[:64], nrow=8).permute(1, 2, 0))
```
