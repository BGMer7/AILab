## 2.3.3 generate data

```python
np.array([1, 2, 3])
```

```python
np.array((1, 2, 3)) # 同上
# array([1, 2, 3])
```

```python
np.array(((1, 2),(1, 2)))
```

```python
np.array(([1, 2],[1, 2])) # 同上
```

```python
np.arange(10) # 10个，不包括10，步长为1
```

```python
np.arange(3, 10, 0.1) # 从3到9，步长为0.1
```

```python
# 从2.0到3.0，生成均匀的5个值，不包括终值3.0
np.linspace(2.0, 3.0, num=5, endpoint=False)
```

```python
# 返回一个6×4的随机数组，浮点型
np.random.randn(6, 4)
```

```python
# 指定范围、指定形状的数组，整型
np.random.randint(3, 7, size=(2, 4))
```

```python
# 创建值为0的数组
np.zeros(6) # 6个浮点0.
```

```python
np.zeros((5, 6), dtype=int) # 5×6整型0
```

```python
np.ones(4) # 同上
```

```python
np.empty(4) # 同上
```

```python
# 创建一份和目标结构相同的0值数组
np.zeros_like(np.arange(6))
```

```python
np.ones_like(np.arange(6)) # 同上
```

```python
np.empty_like(np.arange(6)) # 同上
```

## 2.3.5 array information

```python
n = np.random.randn(6, 4)
n.shape # 数组的形状，返回值是一个元组
```

```python
n = np.zeros(8, dtype=int)
n.shape = (4, 2) # 改变形状，需要刚好可以整除排列
n
```

```python
a = n.reshape((2,4)) # 改变原数组的形状，创建一个新数组
a
```

```python
n.dtype # 数据类型
```

```python
n.ndim # 维度数
```

```python
n.size # 元素数
```

```python
np.array([10, 20, 30, 40])[:3] # 支持类似列表的切片
```

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])
```

```python
a+b # array([11, 22, 33, 44])（矩阵相加）
```

```python
a-1 # array([9, 19, 29, 39])
```

```python
4*np.sin(a)
```

```python
# 以下是一些数学函数的例子，还支持非常多的数学函数
a.max() # 40
```

```python
a.min() # 10
```

```python
a.sum() # 100
```

```python
a.std() # 11.180339887498949
```

```python
a.all() # True
```

```python
a.cumsum() # array([10, 30, 60, 100])
```
